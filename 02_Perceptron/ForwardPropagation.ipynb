```python
import numpy as np
```

```python
def relu(x):
  return np.maximum(0 , x)
# relu use hunxa hamro hidden layer ma
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
# neural network banaune aba
class NeuralNetwork:
  def __init__(self , input_size , hidden_size , output_size):
    self.W1 = np.random.randn(input_size , hidden_size) * 0.01
    self.b1 = np.zeros((1 , hidden_size))
    self.W2 = np.random.randn(hidden_size , output_size) * 0.01
    self.b2 = np.zeros((1, output_size))
  def forward(self, X):
    self.Z1 = np.dot(X,self.W1) + self.b1
    self.A1 = relu(self.Z1)
    self.Z2 = np.dot(self.A1 , self.W2) + self.b2
    self.A2 = sigmoid(self.Z2)
    return self.A2
```

```python
X = np.array([[1.0, 0.5, -0.2], [0.3, -0.1, 0.8]])
```

```python
nn = NeuralNetwork(input_size=3, hidden_size=4, output_size=1)
```

```python
output = nn.forward(X)
```

```python
print("Input Data: \n" , X)
```

```python
print("Output predictions:\n", output)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes (layers)
input_nodes = ['I1', 'I2', 'I3']
hidden_nodes = ['H1', 'H2', 'H3', 'H4']
output_nodes = ['O1']

G.add_nodes_from(input_nodes, layer=0)
G.add_nodes_from(hidden_nodes, layer=1)
G.add_nodes_from(output_nodes, layer=2)

# Add edges (connections)
for i in input_nodes:
    for h in hidden_nodes:
        G.add_edge(i, h)
for h in hidden_nodes:
    G.add_edge(h, output_nodes[0])

# Draw the graph
pos = nx.multipartite_layout(G, subset_key="layer")
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1500, font_size=10, arrows=True)
plt.title("Simple Neural Network (Forward Propagation)")
plt.show()
```
